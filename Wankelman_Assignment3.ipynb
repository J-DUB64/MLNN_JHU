## 2.  Create a Linear Regression model (LIKE WE DID IN CLASS) to fit the data. *Use the example from Lesson 3 and DO NOT USE a library that calculates automatically*. We are expecting 5 coefficients to describe the linear model. 

## After creating the model (finding the coefficients), calculate a new column $y_p=\Sigma\beta_n\cdot x_n$ 

```python
n = len(x1)
```

```python
m1_numer = (np.sum(x1) * np.sum(y) / n) - np.sum(x1 * y)
m1_denom = (np.sum(x1) * np.sum(x1) / n) - np.sum(x1 * x1)
m1 = m1_numer / m1_denom

# Calculation of intercept (b)
b = (np.sum(x1 * x1) * np.sum(y) - np.sum(x1) * np.sum(x1 * y)) / (n * np.sum(x1 * x1) - np.sum(x1) * np.sum(x1))

m1, b
```

```python
m2_numer = (np.sum(x2) * np.sum(y) / n) - np.sum(x2 * y)
m2_denom = (np.sum(x2) * np.sum(x2) / n) - np.sum(x2 * x2)
m2 = m2_numer / m2_denom
b2 = (np.sum(x2 * x2) * np.sum(y) - np.sum(x2) * np.sum(x2 * y)) / (n * np.sum(x2 * x2) - np.sum(x2) * np.sum(x2))

m2, b2
```

```python
m3_numer = (np.sum(x3) * np.sum(y) / n) - np.sum(x3 * y)
m3_denom = (np.sum(x3) * np.sum(x3) / n) - np.sum(x3 * x3)
m3 = m3_numer / m3_denom
b3 = (np.sum(x3 * x3) * np.sum(y) - np.sum(x3) * np.sum(x3 * y)) / (n * np.sum(x3 * x3) - np.sum(x3) * np.sum(x3))

m3, b3
```

```python
m4_numer = (np.sum(x4) * np.sum(y) / n) - np.sum(x4 * y)
m4_denom = (np.sum(x4) * np.sum(x4) / n) - np.sum(x4 * x4)
m4 = m4_numer / m4_denom
b4 = (np.sum(x4 * x4) * np.sum(y) - np.sum(x4) * np.sum(x4 * y)) / (n * np.sum(x4 * x4) - np.sum(x4) * np.sum(x4))

m4, b4
```

```python
X = np.vstack([x1, x2, x3, x4, np.ones(n)]).T
```

```python
Y = np.linspace(0, 1, n) + np.random.rand(n) - 1
```

```python
plt.scatter(X.T[0], Y)
```

```python
plt.scatter(X.T[1], Y)
```

```python
plt.scatter(X.T[2], Y)
```

```python
plt.scatter(X.T[3], Y)
```

```python
left = np.linalg.inv(np.dot(X.T, X))
```

```python
right = np.dot(Y.T, X)
```

```python
np.dot(left, right)
```

```python
beta = np.linalg.lstsq(X, Y)[0]
beta
```

## 3. Plot the model's prediction as a different color on top of the scatter plot from Q1 in 2D for all 4 of the dimensions ($x_1 \rightarrow y_p, x_2 \rightarrow y_p, x_3 \rightarrow y_p, x_4 \rightarrow y_p$) 

```python
pred = np.dot(X, beta)
```
