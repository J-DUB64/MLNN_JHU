```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD  #Stochastic Gradient Descent

import numpy as np
# fix random seed for reproducibility
np.random.seed(7)

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
plt.scatter(*zip(*X), c=y)
```

```python
model = Sequential()

model.add(Dense(8, input_dim = 2, activation = 'tanh'))  #sigmoid, relu
# model.add(Dense(2, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# model.add(Dense(1,input_dim=2, activation='sigmoid'))


sgd = SGD(learning_rate = 0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')

model.fit(X, y, batch_size = 2, epochs = 400) #160/4 = 40 per epoch
print(model.predict(X).reshape(4 * n))

# evaluate the model
scores = model.evaluate(X, y)
```

```python
def create_and_train_model(layers):
    model = Sequential()
    model.add(Dense(2, input_dim = 2, activation = 'tanh'))
    for _ in range(layers - 1):
        model.add(Dense(2, activation = 'tanh'))
    model.add(Dense(1, activation = 'sigmoid'))
    
    sgd = SGD(learning_rate = 0.1)
    model.compile(loss = 'binary_crossentropy', optimizer = sgd)
    
    history = model.fit(X, y, batch_size = 2, epochs = 400, verbose = 1)
    return history.history['loss']

    scores = model.evaluate(X, y)
    print(f"Model with {layers} layers evaluation score: {scores}")
    
    return history.history['loss'], model


layer_counts = [2, 3, 4, 5]
results = []
models = []

for layers in layer_counts:
    loss = create_and_train_model(layers)
    results.append((layers, loss))
    models.append(model)

```

```python
for i, model in enumerate(models):
    print(f"Predictions for model with {layer_counts[i]} layers:")
    print(model.predict(X).reshape(4 * n))
```

```python
layer_counts = [pair[0] for pair in results]
average_losses = [np.mean(pair[1]) for pair in results]

plt.figure(figsize=(8, 5))
plt.plot(layer_counts, average_losses, marker='o', linestyle='-', color='b')

plt.title('Average Loss vs. Number of Layers')
plt.xlabel('Number of Layers')
plt.ylabel('Average Loss')
plt.grid(True)
plt.xticks(layer_counts) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for layers, loss in results:
    plt.plot(loss, label=f'{layers} layers')

plt.title('Model Loss vs. Epochs for Different Layer Counts')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## Using Diabetes data 

http://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data

1. Number of times pregnant 
2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test 
3. Diastolic blood pressure (mm Hg) 
4. Triceps skin fold thickness (mm) 
5. 2-Hour serum insulin (mu U/ml) 
6. Body mass index (weight in kg/(height in m)^2) 
7. Diabetes pedigree function 
8. Age (years) 
9. Class variable (0 or 1) 

```python
import pandas as pd
import numpy as np

# load pima indians dataset
dataset = np.genfromtxt("diabetes.csv", delimiter=",", skip_header=1)
```

```python
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(16, input_dim=8, activation='tanh'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, epochs=1000, batch_size=10)
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
