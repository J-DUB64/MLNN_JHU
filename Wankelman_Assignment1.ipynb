```python
import math

radius = float(input("Enter the radius of the circle: "))
area = math.pi * radius ** 2
print(f"The area of the circle with radius {radius} is {area:.2f}")
```

```python
max_number = int(input("Enter the maximum number: "))
print("Even numbers and their squares up to", max_number)
for i in range(max_number + 1):
    if i % 2 == 0:
        print(f"{i} squared is {i**2}")
```

```python
data = {'Name': ['Joe', 'Suly', 'Ayden', 'Ava'],
        'Age': [42, 32, 18, 14]}
df = pd.DataFrame(data)
print("\nPandas DataFrame:\n", df)


plt.figure(figsize=(8,4))
plt.bar(df['Name'], df['Age'], color='blue')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Bar Chart of Ages')
plt.show()
```

```python
mean_age = df['Age'].mean()
median_age = df['Age'].median()

print("\nMean of Ages:", mean_age)
print("Median of Ages:", median_age)
```
